Generate Powergrids (Problems) and execute brute force

```python
# Hyperparameters
num_busses = 50 # generators and consumers

min_generators, max_generators = 1, 15
generators_step = 1

min_load_factor, max_load_factor = 0.1, 1
load_factor_step = 0.1

num_instances_per_problem_size = 30
brute_force_data_directory = 'data/brute_force/second_run'
```

```python
import numpy as np
num_generators = np.arange(min_generators, max_generators + generators_step, generators_step)
load_factors = np.arange(min_load_factor, max_load_factor + load_factor_step, load_factor_step)
instances = list(range(num_instances_per_problem_size))
```

```python
np.random.seed(1234) # set global seed for reproducibility
```

```python
from brute_force import brute_force
from datetime import datetime
from powergrid import PowerGrid

now = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')

def execute_brute_force(n: int, num_generators: int, load_factor: float, num_instance: int) -> None:
    grid = PowerGrid.random(n=n, num_generators=num_generators)
    print("Num Generators:", num_generators)

    num_branches = sum(grid.num_outgoing_branches())
    c = np.random.randint(1, 1001, size=num_branches)

    file_name = f"{brute_force_data_directory}/candidate_space_{2**num_generators}_candidates_{n}_busses_{round(load_factor, 1)}_load_factor_{now}_{num_instance}.csv"
    best_candidate, best_loss, best_dis = brute_force(
        grid,
        bounds=[(0, 1)] * num_generators,
        c=c,
        load_factor=load_factor,
        write_to_file=file_name,
        verbose=0
    )

    print("Best candidate:", best_candidate)
    print("Best loss:", best_loss)
    print("Best discrepancy", best_dis)
```

```python
import itertools

cartesian_product = list(itertools.product(num_generators, load_factors, instances))
```

```python
print(f"Executing {len(cartesian_product)} tasks")
```

```python
from joblib import Parallel, delayed

Parallel(n_jobs=-1, verbose=11)(
    delayed(execute_brute_force)(n=num_busses, num_generators=num_generators, load_factor=load_factor, num_instance=num_instance)
    for num_generators, load_factor, num_instance in cartesian_product
)
```
