Generate Powergrids (Problems) and execute brute force

```python
# Hyperparameters
num_busses = 50 # generators and consumers

min_generators, max_generators = 1, 20
generators_step = 1

min_load_factor, max_load_factor = 0.1, 1
load_factor_step = 0.1

num_instances_per_problem_size = 10
brute_force_data_directory = 'data/brute_force/test'
```

```python
import numpy as np
num_generators = np.arange(min_generators, max_generators + generators_step, generators_step)
load_factors = np.arange(min_load_factor, max_load_factor + load_factor_step, load_factor_step)
seeds = np.arange(1, num_instances_per_problem_size + 1, step=1)
```

```python
from brute_force import brute_force
from datetime import datetime
from powergrid import PowerGrid


def execute_brute_force(n: int, num_generators: int, load_factor: float, seed: int) -> None:
    np.random.seed(seed)
    grid = PowerGrid.random(n=n, num_generators=num_generators, seed=seed)
    print("Num Generators:", num_generators)

    rng = np.random.default_rng(seed)
    num_branches = sum(grid.num_outgoing_branches())
    c = rng.integers(1, 1001, size=num_branches)

    file_name = f"test/candidate_space_{2**num_generators}_candidates_{n}_busses_{round(load_factor, 1)}_load_factor_{seed}_seed_{datetime.now().strftime('%Y-%m-%d_%H-%M-%S')}.csv"
    best_candidate, best_loss, best_dis = brute_force(
        grid,
        bounds=[(0, 1)] * num_generators,
        c=c,
        load_factor=load_factor,
        write_to_file=file_name,
        verbose=1
    )

    print("Best candidate:", best_candidate)
    print("Best loss:", best_loss)
    print("Best discrepancy", best_dis)
```

```python
import itertools

cartesian_product = list(itertools.product(num_generators, load_factors, seeds))
```

```python
print(f"Executing {len(cartesian_product)} tasks")
```

```python
from joblib import Parallel, delayed

Parallel(n_jobs=-1, verbose=11)(
    delayed(execute_brute_force)(n=num_busses, num_generators=num_generators, load_factor=load_factor, seed=seed)
    for num_generators, load_factor, seed in cartesian_product
)
```
